# **1) Load Data**

```python
import pandas as pd
df=pd.read_excel('https://archive.ics.uci.edu/ml/machine-learning-databases/00477/Real%20estate%20valuation%20data%20set.xlsx')
df
```

# **2) Data Preparation**

## **2-1) Split Loaded Data to X and Y**

```python
y=df['Y house price of unit area']
y
```

```python
x=df.drop('Y house price of unit area',axis=1)
x
```

## **2-2) Split Data for Training & Test**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test =train_test_split(x,y,test_size=0.2,random_state=59)
```

```python
X_train
```

```python
X_test
```

# **3) Model Building**

## **3-1) Lineae Regression**

### **3-1-1) Training The Model**

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(X_train,y_train)
```

### **3-1-2)Apply The Model To Make Prediction**

```python
y_lr_train_pred=lr_model.predict(X_train)
y_lr_test_pred=lr_model.predict(X_test)
```

### **3-1-3)Evaluate Model Preformance**


```python
from sklearn.metrics import mean_squared_error , r2_score
lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred     )
lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_results=pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','MSE (Train)','R2 (Train)',' MSE (Test) ',' R2 (Test)']
lr_results
```

# **4) Data Visualization**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,5))
plt.scatter(x=y_train,y=y_lr_train_pred, c='#ff7f0e')
z=np.polyfit(y_train,y_lr_train_pred,1)
p=np.poly1d(z)
plt.plot(y_train,p(y_train),'#000000')
plt.plot()
plt.ylabel('predicted Prices')
plt.xlabel('Experimental Prices')
plt.title('Data Visualization')

```
