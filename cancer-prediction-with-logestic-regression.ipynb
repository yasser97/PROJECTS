```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

**Import The Data You Will Work On It**

```python
data=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
data.head()
```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace=True)
data.head()
```

```python
data.diagnosis=[1 if value=="M"else 0 for value in data.diagnosis]
```

```python
data.head()
```

**Split the data into target value and predictors**

```python
y=data["diagnosis"]
X=data.drop(["diagnosis"],axis=1)
```

```python
X
```

**Normalize Your Data**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

**Split the data into training set and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.4,random_state=42)
```

**Train Your Model**

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
y_pred
```

**Evaluation of Your Model**

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print(f"accuracy:{accuracy: .3f}")
```

```python
from sklearn.metrics import classification_report
print( classification_report(y_test,y_pred))
```

**Visulaize Your Data**

```python
plt.figure(figsize=(8, 6)) 
sns.scatterplot(x=y_test, y=y_pred, hue=y_pred, palette="coolwarm", alpha=0.7) 
plt.ylabel('Prediction')
plt.title('Our test data')
```
